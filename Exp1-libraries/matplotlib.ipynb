```python
import matplotlib.pyplot as plt
import numpy as np
```

Line plot

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 30, 20, 50, 40])
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='+')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [20, 10, 30, 40, 35]

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, color="blue")
plt.show()
```

```python
plt.plot(x,y, color="blue", linewidth=5, linestyle="dashdot")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [20, 10, 30, 40, 35]
y2 = [20, 30, 30, 10, 20]

plt.title("Weather")
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.plot(x, y1, label="max")
plt.plot(x, y2, label="min")
plt.legend(loc="upper right", shadow="true", fontsize="large")  # default="best"
plt.grid()
```

Bar plot

```python
company = ["google", "amazon", "facebook"]
revenue = [90,100,50]
plt.bar(company,revenue)
plt.show()
```

Scatter plot

```python
x = [1,3,6,7,1,10]
y = [20,30,10,5,10,15]
plt.xlabel("X label")
plt.ylabel("Y label")
plt.scatter(x,y)
plt.show()
```

Histogram plot

```python
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

Pie chart

```python
arr = np.array([35, 45, 5, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["yellow", "hotpink", "b", "#4CAF50"]
myexplode= [0.2, 0, 0, 0]
plt.pie(arr, labels=mylabels, colors=mycolors, explode=myexplode, radius=1.1)
plt.legend(title="Four fruits",loc="upper right")
plt.show()
```

Subplots

```python
fig, axes = plt.subplots(nrows=2,ncols=3, figsize=(10,5))
```

```python
axes[0][0].plot([1,3,5], [5,10,15], c="green")
axes[0][1].plot([1,8,10], [20,1,5], c="blue")
fig
```

```python
plt.subplot(1,2,1)
plt.plot([1,3,5], [5,10,15], c="green")
plt.subplot(1,2,2)
plt.plot([1,8,10], [20,1,5], c="blue")
plt.show()
```
