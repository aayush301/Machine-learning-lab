```python
import seaborn as sns
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from kmeans import KMeans
```

```python
centers = 5
X_train, true_labels = make_blobs(n_samples=100, centers=centers, random_state=42)
X_train = StandardScaler().fit_transform(X_train)
sns.scatterplot(x=X_train[:,0], y=X_train[:,1], hue=true_labels, palette="deep", legend=None)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
kmeans = KMeans(n_clusters=centers)
kmeans.fit(X_train)
```

```python
sns.scatterplot(x=X_train[:,0], y=X_train[:,1], hue=true_labels, palette="deep", legend=None)
class_centers, class_ = kmeans.predict(X_train)
x_values = np.array(class_centers)[:,0]
y_values = np.array(class_centers)[:,1]
plt.plot(x_values, y_values, '+', markersize=10)
plt.show()
```
